```python
import torch

def collect_all_image_ids(checkpoint_path):
    # Load the checkpoint
    print(f"Loading {checkpoint_path}")
    storage = torch.load(checkpoint_path)
    
    all_image_ids = set()  # Using set to avoid duplicates
    total_image_ids = []
    # Iterate through all dimensions of the storage
    for strategy in storage.keys():
        for interval in storage[strategy].keys():
            for sublayer in storage[strategy][interval].keys():
                image_ids_data = storage[strategy][interval][sublayer]['image_ids']
                
                # Iterate through layers
                for layer_data in image_ids_data:
                    # Iterate through neurons
                    for neuron_data in layer_data:
                        # Add all image IDs to the set
                        all_image_ids.update(neuron_data)
                        total_image_ids.append(neuron_data)
    
    # Convert set to sorted list
    all_image_ids = sorted(list(all_image_ids))
    print(f"Total unique image IDs found: {len(all_image_ids)}")
    
    return all_image_ids, total_image_ids
```
